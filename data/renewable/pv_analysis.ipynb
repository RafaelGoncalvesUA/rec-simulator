```python
import pandas as pd
import datetime
```

```python
pv = pd.read_csv('renewable5.csv.gz', compression='gzip')
pv.columns = ['step', 'pv']
pv.drop(['step'], axis=1, inplace=True)

pv["time"] = pd.date_range(start="2014-01-01", periods=len(pv), freq="15min")

pv.head()
```

```python
pv.iloc[:300].plot(x='time', y='pv', title='PV Generation', figsize=(12, 6), grid=True)
```
